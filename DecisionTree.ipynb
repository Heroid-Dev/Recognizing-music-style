```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# خواندن داده‌ها از فایل CSV و ایجاد دیتافریم
my_data = pd.read_csv("./MusicDataSet.csv", delimiter=",")

# حذف ستون "Unnamed: 0" از دیتافریم
my_data = my_data.drop("Unnamed: 0", axis=1)

# نمایش پنج نمونه اول از دیتافریم
print(my_data[0:5])
```

```python
# حذف ستون‌های "File Name" و "Label" و ایجاد ماتریس ویژگی‌ها (X)
X = my_data.drop(["File Name", "Label"], axis=1)

# جداگانه ذخیره کردن ستون "Label" به عنوان برچسب‌ها (y)
y = my_data["Label"]

# نمایش پنج نمونه اول از ماتریس ویژگی‌ها (X)
print(X[:5])
```

```python
# نرمال‌سازی داده‌های ویژگی با استفاده از StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# انجام مقیاس‌دهی به داده‌های ویژگی و ذخیره نتیجه در ماتریس X_scaled
X_scaled = scaler.fit_transform(X)
```

```python
# داده‌های آموزشی و آزمون از داده‌های ویژگی نرمال‌شده جداگانه جدا می‌شوند
# test_size: نسبت داده‌های آزمون به کل داده‌ها
# random_state: تعداد تصادفی‌سازی برای تقسیم داده‌ها
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python

# criterion: معیار مورد استفاده برای تقسیم‌بندی (در اینجا انتروپی انتخاب شده است)
# max_depth: حداکثر عمق مجاز برای درخت تصمیم (تا چه عمقی تصمیم‌ها به صورت بازگشتی تا خروجی‌ها ادامه پیدا کنند)
# random_state: تعداد تصادفی‌سازی برای ایجاد تصمیم‌گیری یکتا
clf = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=42)

# آموزش مدل با داده‌های آموزشی
clf.fit(X_train, y_train)
```

```python
# پیش‌بینی برچسب‌ها با استفاده از مدل درخت تصمیمی بر روی داده‌های تست
predTree = clf.predict(X_test)
```

```python
print (predTree [0:5])
print (y_test [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
# محاسبه و چاپ دقت مدل درخت تصمیمی بر روی داده‌های تست
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```
